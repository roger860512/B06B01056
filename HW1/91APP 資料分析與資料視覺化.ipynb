```python
import pandas as pd
```

```python
datasets = pd.read_csv('/Users/roger/91App/PromotionConditions.csv')
datasets
#此資料集共有8個欄位，分別是'Promotionld' , 'PromotionStartDateTime' , 'PromotionEndDateTime' , 'PromotionValidFlag' , 'PromotionCondition_TotalPrice' ,
#                                         'PromotionCondition_DiscountRate' , 'PromotionCondition_TotalQty' , 'PromotionCondition_DiscountPrice' , 'DiscountType'
#首先，我們要做的是分析各個欄位，必選定所要視覺化之資料，藉由視覺化我們可以更具體的看出資料所代表的含義，也能更好的完成資料分析。
#最後，我們能得出我們要的結果，，更好地達成目標。
```

```python
#分析欄位：
#首先，我們先選定要分類的資料以及視覺化的目標
#選擇'PromotionCondition_TotalPrice' ,'PromotionCondition_DiscountRate' , 'PromotionCondition_TotalQty' 作為所要分析之欄位
#我們可以藉由此三欄位分析Total Price & Discount Price 與 TotalQty'之間的關係，藉由樸素貝葉斯分類，並將其視覺化，以得出三者間的關係
X = datasets.iloc[:, [4, 7]].values
y = datasets.iloc[:, 6].values
```

```python
#將X與y分為訓練集與測試集。
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Fitting Naive Bayes to the training set
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
#Predicting the Test set （訓練集）results
#for what?
#
y_pred = classifier.predict(X_test)
```

```python
#Making Confusion Matrix 
#for what？
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
import numpy as np
# Visualising the Training set results
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
    
```

```python
#放上標題、Xy軸分別代表之意思以及圖示表
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green','blue','black')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green','blue','black'))(i), label = j)
plt.title('Naive Bayes (Training set)')
plt.xlabel('TotalPrice')
plt.ylabel('DiscountPrice')
plt.legend()
plt.show()
#觀察此圖，我們可以得出當DiscountRate 的值不等於0 時，DiscountPrice 的大小才會與其有關係
#而DiscountRate與TotalPrice就更淺顯易懂了，我們可以看出TotalPrice的大小對DiscountRate是決定性的
#(當TotalPrice大於某值DiscountRate必為0)
```
